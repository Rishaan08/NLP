```python
import pandas as pd
messages = pd.read_csv('SMSSpamCollection.txt',sep='\t',names=['label','message'])
```

```python
from nltk.stem import WordNetLemmatizer
lemma = WordNetLemmatizer()
```

```python
import re
from nltk.corpus import stopwords
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [lemma.lemmatize(word) for word in review]
    review = ' '.join(review)
    corpus.append(review)
```

```python
[[i,j,k] for i,j,k in zip(list(map(len,corpus)),corpus,messages['message']) if i < 1]
```

```python
corpus
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
words = []
for sent in corpus:
    sent_token = sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
words
```

```python
#Train the model from scratch
model = gensim.models.Word2Vec(words)

#To get all the vocabulary
model.wv.index_to_key
```

```python
import numpy as np
def avg_word2vec(doc):
    #Remove out of vocabulary words
    #sent = [word for word in doc if word in model.wv.index_to_key]
    #print(sent)

    return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key],axis=0)
```

```python
from tqdm import tqdm
X = []
for i in tqdm(range(len(words))):
    X.append(avg_word2vec(words[i]))
```

```python
#Dependent Feature
y = messages[list(map(lambda x: len(x)>0,corpus))]
y = pd.get_dummies(y['label'])
y = y.iloc[:,0].values
```

```python
y.shape
```

```python
rows = []

# Iterate through X and append reshaped rows to the list
for i in range(0, len(X)):
    rows.append(pd.DataFrame(X[i].reshape(1, -1)))

# Use pd.concat to concatenate all DataFrame rows
df = pd.concat(rows, ignore_index=True)
```

```python
df.shape
```

```python
df.head()
```

```python
#Independent Feature
X = df 
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
spam_detect_model = RandomForestClassifier().fit(X_train,y_train)
```

```python
y_pred = spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
