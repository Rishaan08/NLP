```python
import pandas as pd
messages = pd.read_csv('SMSSpamCollection.txt',sep='\t',names=['label','message'])
```

```python
messages
```

```python
 #Data Cleaning and Preprocessing
import re, nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lemma = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0,len(messages)):
    review = re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [lemma.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

# Bag of Words

```python
#Output Features
y = pd.get_dummies(messages['label'])
y = y.iloc[:,1].values
```

```python
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
y = le1.fit_transform(messages['label'])
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(corpus,y,test_size=0.20)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500,binary=True,ngram_range=(1,2))
```

```python
X_train = cv.fit_transform(X_train).toarray()
X_test = cv.transform(X_test).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000,
            formatter=dict(float=lambda x: '%.3g' % x))

X_train
```

```python
cv.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train,y_train)
```

```python
spam_detect_model
```

```python
y_pred = spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

# TF-IDF

```python
#Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(corpus,y,test_size=0.20)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=2500,ngram_range=(1,2))
X_train = tfidf.fit_transform(X_train).toarray()
X_test = tfidf.transform(X_test).toarray()
```

```python
X_train
```

```python
tfidf.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_tfidf_model = MultinomialNB().fit(X_train,y_train)
```

```python
spam_tfidf_model
```

```python
y_pred1 = spam_tfidf_model.predict(X_test)
```

```python
accuracy_score(y_test,y_pred1)
```

```python
print(classification_report(y_test,y_pred1))
```
